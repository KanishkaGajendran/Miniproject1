**SecureCheck: A Python-SQL Digital Ledger for Police Post Logs**

```python
# A Python-SQL Digital Ledger for Police Post Logs
# Python for Data processing

import pandas as pd
import numpy as np
# read the file
df = pd.read_excel("D:/Data _Science/Projects/Project 1/traffic_stops.xlsx")
```

```python
df
```

```python
#Remove the column contain all missing values
df.dropna(axis=1,how='all',inplace=True)

#Handle the NAN values
df.fillna({'search_type':'Nothing'},inplace=True)
```

```python
df
```

```python
#convert stop_time datatype
df['stop_time']=pd.to_datetime(df['stop_time'], format= '%H:%M:%S')
```

```python
df.dtypes
```

```python
import pymysql
```

```python
myconnection= pymysql.connect(host="localhost",user="root",password="Gaja!Kani2022")

cur=myconnection.cursor()
cur.execute("create database DS_Secure_check")
cur.execute("use DS_Secure_check")
#cur.execute("drop database DS_Secure_check")
```

```python
# Create a table
cur.execute("""CREATE TABLE IF NOT EXISTS check_post_log(
s_no INT AUTO_INCREMENT PRIMARY KEY,
stop_date DATE,
stop_time TIME,
country_name VARCHAR(20),
driver_gender VARCHAR(10),
driver_age INT,
driver_race VARCHAR(50),
violation VARCHAR(50),
search_conducted BOOLEAN,
search_type VARCHAR(50),
stop_outcome VARCHAR(50),
is_arrested BOOLEAN,
stop_duration VARCHAR(50),
drugs_related_stop BOOLEAN,
vehicle_number VARCHAR(20)
)""")

myconnection.commit()
print("Table created sucessfully")
```

```python
query=""" INSERT INTO check_post_log(stop_date,stop_time,country_name,driver_gender,driver_age,
                                     driver_race,violation,search_conducted,search_type,stop_outcome,
                                     is_arrested,stop_duration,drugs_related_stop,vehicle_number)
                                     values(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""
for _, row in df.iterrows():
    cur.execute(query,(row['stop_date'],
                          row['stop_time'],
                          row.get('country_name'),
                          row.get('driver_gender'),
                          int(row.get('driver_age'))if row.get('driver_age')is not None else None,
                          row.get('driver_race'),
                          row.get('violation'),
                          bool(row.get('search_conducted'))if row.get('search_conducted')is not None else None,
                          row.get('search_type'),
                          row.get('stop_outcome'),
                          bool(row.get('is_arrested'))if row.get('is_arrested')is not None else None,
                          row.get('stop_duration'),
                          bool(row.get('drugs_related_stop'))if row.get('drugs_related_stop')is not None else None,
                          row.get('vehicle_number', f'VEH{row.name}')
                         ))
    myconnection.commit()
print("Inserted data successfully")
```
